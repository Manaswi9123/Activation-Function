```python
import math
def sig(x):
    return 1/(1+math.exp(-x))
```

```python
sig(100)
```

```python
def tanh(x):
    return (math.exp(x)-math.exp(-x))/(math.exp(x)+math.exp(-x))
```

```python
tanh(-56)
```

```python
def reLU(x):
    return max(0,x)
```

```python
reLU(-9)
```

```python
def leaky_reLU(x):
    return max(0.1*x,x)
```

```python
leaky_reLU(-10)
```
